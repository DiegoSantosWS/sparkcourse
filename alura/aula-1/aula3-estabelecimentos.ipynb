```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .master('local[*]') \
    .appName("Carregando dados estabelecimentos") \
    .getOrCreate()
```

```python
# imports and path
from pyspark.sql.types import DoubleType, StringType
from pyspark.sql import functions as func

path = "../datasets/alura/01/estabelecimentos/"
```

```python
estabes = spark.read.csv(path, sep=';', inferSchema=True)
estabes.count()
```

```python
estabes.limit(5).toPandas()
```

```python
estabes.printSchema()
```

```python
estabsColNames = ['cnpj_basico', 'cnpj_ordem', 'cnpj_dv', 'identificador_matriz_filial', 'nome_fantasia', 'situacao_cadastral', 'data_situacao_cadastral', 'motivo_situacao_cadastral', 'nome_da_cidade_no_exterior', 'pais', 'data_de_inicio_atividade', 'cnae_fiscal_principal', 'cnae_fiscal_secundaria', 'tipo_de_logradouro', 'logradouro', 'numero', 'complemento', 'bairro', 'cep', 'uf', 'municipio', 'ddd_1', 'telefone_1', 'ddd_2', 'telefone_2', 'ddd_do_fax', 'fax', 'correio_eletronico', 'situacao_especial', 'data_da_situacao_especial']
```

```python
for idx, name in enumerate(estabsColNames):
    estabes = estabes.withColumnRenamed(f"_c{idx}", name)
```

```python
estabes.limit(5).toPandas()
```

```python
estabes.printSchema()
```

```python
estabes.limit(5).toPandas()
```

```python
estabes = estabes.\
    withColumn(
        'data_situacao_cadastral', func.to_date(estabes.data_situacao_cadastral.cast(StringType()), 'yyyyMMdd')
    ).withColumn(
        'data_de_inicio_atividade', func.to_date(estabes.data_de_inicio_atividade.cast(StringType()), 'yyyyMMdd')
    ).withColumn(
        'data_da_situacao_especial', func.to_date(estabes.data_da_situacao_especial.cast(StringType()), 'yyyyMMdd')
    )
    
estabes.printSchema()
```

```python
estabes.limit(5).toPandas()
```

```python
sel = estabes.select(
    'nome_fantasia', 
    'municipio', 
    func.year('data_de_inicio_atividade').alias("ano_de_inicio_atividade"),
    func.month('data_de_inicio_atividade').alias("mes_de_inicio_atividade"),
)
sel.limit(5).toPandas()
```
